```python
from sklearn.preprocessing import OrdinalEncoder

# Initialize OrdinalEncoder with handling for unknown values
oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

# Apply ordinal encoding to the categorical columns in X_train
X_train[['gender', 'match_type', 'out_kind']] = oe.fit_transform(X_train[['gender', 'match_type', 'out_kind']])

# Display the first few rows of the modified X_train to verify
display(X_train.head())
```

```python
model.fit(X_train, y_train)
```

```python
import numpy as np

def top_11_mae(y_true, y_pred, match_ids):
    """
    Calculates the Mean Absolute Error between the top 11 actual and predicted fantasy scores per match.

    Args:
        y_true: Array-like of actual fantasy scores.
        y_pred: Array-like of predicted fantasy scores.
        match_ids: Array-like of match IDs corresponding to each score.

    Returns:
        The average MAE between the top 11 actual and predicted scores across all matches.
    """
    mae_per_match = []
    unique_match_ids = np.unique(match_ids)

    for match_id in unique_match_ids:
        # Get the indices for the current match
        match_indices = np.where(match_ids == match_id)[0]

        if len(match_indices) < 11:
            # Skip matches with fewer than 11 players
            continue

        y_true_match = y_true[match_indices]
        y_pred_match = y_pred[match_indices]

        # Get the indices of the top 11 actual and predicted scores for this match
        top_11_true_indices = np.argsort(y_true_match)[-11:]
        top_11_pred_indices = np.argsort(y_pred_match)[-11:]

        # Get the actual and predicted scores for the top 11 players in this match
        top_11_true_scores = y_true_match[top_11_true_indices]
        top_11_pred_scores = y_pred_match[top_11_pred_indices]

        # Calculate the Mean Absolute Error for this match
        mae = np.mean(np.abs(top_11_true_scores - top_11_pred_scores))
        mae_per_match.append(mae)

    # Calculate the average MAE across all matches
    average_mae = np.mean(mae_per_match)

    return average_mae
```
